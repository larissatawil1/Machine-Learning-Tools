```python
# Step 1: Load packages & preprocess data into scaled input variables & an output variable
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.utils import plot_model
from pdf2image import convert_from_path
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv("CLV.csv")
data
```

```python
data_select = data[["Income", "Total Claim Amount", "Months Since Last Claim", "EmploymentStatus_Employed", "Customer Lifetime Value"]]
data_select
```

```python
X = data_select.drop(["Customer Lifetime Value"], axis = 1)
y = data['Customer Lifetime Value']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state = 101)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
## Step 2: Run a GridSearch CV on at least 10 possible combinations of hyper parameters 
```

```python
clf = MLPRegressor()

params = {
    "hidden_layer_sizes": [(10,), (20,5,)], 
    "activation" : ['relu','logistic'],
    "alpha" : [0.0001, 0.001, 0.01]}
```

```python
grid = GridSearchCV(clf, params, cv = 10)
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
## Step 3: Train a model with the optimal solution from GridSearch
```

```python
model = MLPRegressor(**grid.best_params_)
model.fit(X_train, y_train)
```

```python
#MLPRegressor(hidden_layer_sizes = (20, 5), 
             activation = "relu")
```

```python
## Step 4: What are the in-sample and out of sample MSEs
```

```python
# In-Sample
y_pred = grid.predict(X_train)
mean_squared_error(y_train, y_pred)
```

```python
# Out-of-Sample
y_pred = grid.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
## Step 5: Build a Keras with the architecture defined by GridSearchCV
```

```python
import keras.models
from keras.optimizers import Adam
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(12, input_dim=X_train.shape[1], activation = 'relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(1, activation = 'softmax'))
```

```python
model.compile(loss='mse', optimizer = Adam(lr=0.01))
model.fit(X_train, y_train, batch_size=32, epochs=100)
```

```python
keras.models.save_model(model, "MLModels")
```
